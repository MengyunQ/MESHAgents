```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE


confounder = ['Sex', 'Age', 'Weight', 'Height', "Alcohol intake","Alcohol intake log10","Systolic blood pressure","Diastolic blood pressure"]

expert_selected = [
        'RVEDV (mL)', 'LVEDV (mL)', 'LVEF (%)', 'LVM (g)', 'RVEF (%)',
        'LAV max (mL)', 'LAEF (%)', 'RAV max (mL)', 'RAEF (%)',
        'AAo max area (mm2)', 'AAo distensibility (10-3 mmHg-1)',
        'DAo max area (mm2)', 'DAo distensibility (10-3 mmHg-1)'
    ]
mesh_discovered = [
        'RVEDV (mL)', 'RVESV (mL)', 'RVSV (mL)', 'LVCO (L/min)',
        'LAV min (mL)', 'LASV (mL)', 'RAV max (mL)', 'RAV min (mL)',
        'AAo max area (mm2)', 'AAo min area (mm2)', 'DAo min area (mm2)',
        'WT_AHA_2 (mm)', 'Ell_4 (%)'
    ]


feature_set = ['LVEDV (mL)', 'LVESV (mL)', 'LVSV (mL)', 'LVEF (%)',
       'LVCO (L/min)', 'LVM (g)', 'RVEDV (mL)', 'RVESV (mL)', 'RVSV (mL)',
       'RVEF (%)', 'LAV max (mL)', 'LAV min (mL)', 'LASV (mL)', 'LAEF (%)',
       'RAV max (mL)', 'RAV min (mL)', 'RASV (mL)', 'RAEF (%)',
       'AAo max area (mm2)', 'AAo min area (mm2)',
       'AAo distensibility (10-3 mmHg-1)', 'DAo max area (mm2)',
       'DAo min area (mm2)', 'DAo distensibility (10-3 mmHg-1)',
       'WT_AHA_1 (mm)', 'WT_AHA_2 (mm)', 'WT_AHA_3 (mm)', 'WT_AHA_4 (mm)',
       'WT_AHA_5 (mm)', 'WT_AHA_6 (mm)', 'WT_AHA_7 (mm)', 'WT_AHA_8 (mm)',
       'WT_AHA_9 (mm)', 'WT_AHA_10 (mm)', 'WT_AHA_11 (mm)', 'WT_AHA_12 (mm)',
       'WT_AHA_13 (mm)', 'WT_AHA_14 (mm)', 'WT_AHA_15 (mm)', 'WT_AHA_16 (mm)',
       'WT_Global (mm)', 'Ecc_AHA_1 (%)', 'Ecc_AHA_2 (%)', 'Ecc_AHA_3 (%)',
       'Ecc_AHA_4 (%)', 'Ecc_AHA_5 (%)', 'Ecc_AHA_6 (%)', 'Ecc_AHA_7 (%)',
       'Ecc_AHA_8 (%)', 'Ecc_AHA_9 (%)', 'Ecc_AHA_10 (%)', 'Ecc_AHA_11 (%)',
       'Ecc_AHA_12 (%)', 'Ecc_AHA_13 (%)', 'Ecc_AHA_14 (%)', 'Ecc_AHA_15 (%)',
       'Ecc_AHA_16 (%)', 'Ecc_Global (%)', 'Err_AHA_1 (%)', 'Err_AHA_2 (%)',
       'Err_AHA_3 (%)', 'Err_AHA_4 (%)', 'Err_AHA_5 (%)', 'Err_AHA_6 (%)',
       'Err_AHA_7 (%)', 'Err_AHA_8 (%)', 'Err_AHA_9 (%)', 'Err_AHA_10 (%)',
       'Err_AHA_11 (%)', 'Err_AHA_12 (%)', 'Err_AHA_13 (%)', 'Err_AHA_14 (%)',
       'Err_AHA_15 (%)', 'Err_AHA_16 (%)', 'Err_Global (%)', 'Ell_1 (%)',
       'Ell_2 (%)', 'Ell_3 (%)', 'Ell_4 (%)', 'Ell_5 (%)', 'Ell_6 (%)',
       'Ell_Global (%)']

target_labels = ['Hypertension', 'High cholesterol', 'Cardiac disease', 'PVD', 'Stroke',
       'Asthma', 'COPD', 'Bronchitis', 'Diabetes', 'Dementia',
       'Depression']
target_labels_renamed = ['Hypertension', 'High cholesterol', 'Cardiac disease', 'PVD', 'Stroke',
       'Asthma', 'COPD', 'Bronchitis', 'Diabetes', 'Dementia',
       'Depression']


data.dropna(subset=target_labels, inplace=True)


imputer = SimpleImputer(strategy='mean')
data[feature_set + confounder] = imputer.fit_transform(data[feature_set + confounder])

feature_sets = {
    'expert_selected': expert_selected + confounder,
    'mesh_discovered': mesh_discovered + confounder
}

metrics = ["AUC", "Recall"]

classifiers = {
    'LDA': LDA()
}
results = {clf: {case: {metric: [] for metric in metrics} for case in feature_sets.keys()} for clf in
           classifiers.keys()}
results['Label'] = target_labels_renamed

for label in target_labels:
    print(f"\nEvaluating target: {label}")
    data_filtered = data[data[label].isin([0, 1])].copy()
    class_counts = data_filtered[label].value_counts()
    num_zeros = class_counts.get(0, 0)
    num_ones = class_counts.get(1, 0)
    min_class_size = min(num_zeros, num_ones)
    
    df_zeros = data_filtered[data_filtered[label] == 0].sample(n=min_class_size, random_state=42)
    df_ones = data_filtered[data_filtered[label] == 1].sample(n=min_class_size, random_state=42)

    balanced_data = pd.concat([df_zeros, df_ones])

    y_balanced = balanced_data[label]
    for clf_name, clf in classifiers.items():
        
        for case, features in feature_sets.items():
            X_balanced = balanced_data[features]
            X_train, X_test, y_train, y_test = train_test_split(
                X_balanced, y_balanced, test_size=0.3, random_state=42
            )
            clf.fit(X_train, y_train)

            y_prob = clf.predict_proba(X_test)[:, 1]
            y_pred = clf.predict(X_test)
            auc = roc_auc_score(y_test, y_prob)
            recall = recall_score(y_test, y_pred)
            accuracy = accuracy_score(y_test, y_pred)
            precision = precision_score(y_test, y_pred, zero_division=0)
            f1 = f1_score(y_test, y_pred)

            results[clf_name][case]['AUC'].append(auc)
            results[clf_name][case]['Recall'].append(recall)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

selected_feature_set = 'mesh_discovered'

target_labels = ['Hypertension', 'High cholesterol', 'Cardiac disease', 'Stroke',
                 'Asthma', 'COPD', 'Diabetes', 'Depression']
selected_classifiers = ['LDA']

fig, ax = plt.subplots(figsize=(6, 6))

for clf_name in selected_classifiers:
    clf = classifiers[clf_name]

    for label in target_labels:
        data_filtered = data[data[label].isin([0, 1])].copy()
        class_counts = data_filtered[label].value_counts()
        min_class_size = min(class_counts.get(0, 0), class_counts.get(1, 0))

        df_zeros = data_filtered[data_filtered[label] == 0].sample(n=min_class_size, random_state=42)
        df_ones = data_filtered[data_filtered[label] == 1].sample(n=min_class_size, random_state=42)
        balanced_data = pd.concat([df_zeros, df_ones])

        X = balanced_data[feature_sets[selected_feature_set]]
        y = balanced_data[label]

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

        clf.fit(X_train, y_train)

        y_prob = clf.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        roc_auc = auc(fpr, tpr)

        ax.plot(fpr, tpr, label=f"{label} (AUC={roc_auc:.3f})")

ax.plot([0, 1], [0, 1], color="gray", linestyle="--")

ax.set_title(
    "ROC Curves - LDA\n"
    "(Model: $\\mathtt{MESH}$)",
    fontsize=16, fontweight='bold'
)

ax.set_xlabel("False Positive Rate", fontsize=14, weight='bold')
ax.set_ylabel("True Positive Rate", fontsize=14, weight='bold')
ax.tick_params(axis='both', which='major', labelsize=12)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontweight('bold')

ax.legend(loc="lower right", prop={"size": 10, "weight": "bold"})
ax.grid(True)

plt.tight_layout()
plt.savefig("roc_curve.pdf", dpi=300, bbox_inches="tight")

plt.show()
```
